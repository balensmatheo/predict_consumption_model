## **Préparation des données**

````python
# Extraction des features temporelles
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['weekday'] = df['Date'].dt.weekday
df['hour'] = df['Heure'].str[:2].astype(int) 

# Variable saison
df['season'] = df['month']%12 // 3 + 1  # 1: hiver, 2: printemps, 3: été, 4: automne
````

````python
# Remplissage des valeurs numériques par interpolation
df.interpolate(method='linear', inplace=True)

# Traitement des colonnes 'ND' : convertir en NaN puis 0
for col in ["Stockage batterie (MW)",
            "Déstockage batterie (MW)",
               "Eolien terrestre (MW)",
                 "Eolien offshore (MW)",
                   "Ech. comm. Angleterre (MW)",
                     "Ech. comm. Angleterre (MW)",
                     "Ech. comm. Allemagne-Belgique (MW)",
                     "Ech. comm. Italie (MW)",
                     "Ech. comm. Espagne (MW)",
                     "Ech. comm. Suisse (MW)"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col].replace('ND', pd.NA), errors='coerce').fillna(0)
````

# Analyse de la répartition des données  

Recherche d'eventuelles valeurs abbérantes (Outliers)

````python
import seaborn as sns

# Sélection des colonnes à analyser (tu adaptes selon tes colonnes réelles)
production_columns = [
    'Fioul (MW)', 'Charbon (MW)', 'Gaz (MW)', 'Nucléaire (MW)',
    'Eolien (MW)', 'Solaire (MW)', 'Hydraulique (MW)', 'Bioénergies (MW)'
]

exchange_columns = [col for col in X.columns if 'Ech.' in col]

# Fonction générique pour afficher Histogramme + Boxplot côte à côte
def plot_distribution(data, col_name):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Histogramme avec courbe de densité (KDE)
    sns.histplot(data[col_name], bins=50, kde=True, ax=axes[0], color='steelblue')
    axes[0].set_title(f'Histogramme de {col_name}')
    axes[0].set_xlabel(col_name)

    # Boxplot pour visualiser les outliers
    sns.boxplot(x=data[col_name], ax=axes[1], color='lightcoral')
    axes[1].set_title(f'Boxplot de {col_name}')

    plt.tight_layout()
    plt.show()

# Affichage pour les colonnes de production
for col in production_columns:
    if col in X.columns:
        plot_distribution(X, col)

# Affichage pour les colonnes d'échanges
for col in exchange_columns:
    plot_distribution(X, col)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error
import xgboost as xgb
import numpy as np

# Split Train/Test sans mélanger (important sur séries temporelles)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

# Modèle XGBoost simple
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=6)
model.fit(X_train, y_train)

# Prédictions
y_pred = model.predict(X_test)

# Évaluation des performances
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f"R² Score : {r2:.3f}")
print(f"RMSE : {rmse:.2f} MW")
print(f"MAPE : {mape:.2%}")
````

````python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error
import numpy as np

tscv = TimeSeriesSplit(n_splits=5)

r2_scores = []
rmse_scores = []
mape_scores = []

for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=6)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    r2_scores.append(r2_score(y_test, y_pred))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    mape_scores.append(mean_absolute_percentage_error(y_test, y_pred))

print(f"R2 mean: {np.mean(r2_scores):.3f}, RMSE mean: {np.mean(rmse_scores):.2f} MW, MAPE mean: {np.mean(mape_scores):.2%}")
````
